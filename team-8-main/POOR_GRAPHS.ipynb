```python
pf.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
pf = pf.rename(columns={'Country' : 'Area'})
```

```python
ad = pd.read_csv('../shared_data_read_only/Agriculture Dataset.csv')
ad = ad[['Area', 'Year', 'Population']]
ad["rank"] = ad.groupby("Area")["Year"].rank("max", ascending=False)
ad = ad[ad["rank"] == 1]
ad.drop(["rank"], axis=1, inplace=True)
ad
```

```python
new = pd.merge(df, pf, on=['Area'], how="outer")
new = pd.merge(new, ad, on=['Area'], how="outer")
```

```python
new = new.dropna(subset=['Area Code', 'Year_x', 'Number of People Working In Agriculture', 'Year_y', 'Percent of Population Below Poverty Line','Year','Population'])
```

```python
new = new.rename(columns={'Year_x' : 'Year_agri', 'Year_y' : 'Year_poor', 'Year' : 'Year_pop', 'Number of People Working In Agriculture' : 'aggies', 'Percent of Population Below Poverty Line': 'Poor Percent'})
new['ag_rate'] = new['aggies'] / new['Population']
```

```python
new['Poor Percent'] = new['Poor Percent'].str.rstrip("%").astype(float)/100
```

```python
new['PPeeps'] = new['Poor Percent'] * new['Population'] 
new
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(new.corr(), cbar=True, annot=True, cmap = 'Blues');
```

GDP per Capita --> GDP / Population

```python
ad_2 = pd.read_csv('../shared_data_read_only/Agriculture Dataset.csv')
ad_2 = ad_2[['Area', 'Year', 'GDP  (2015 US$, mil)']]
ad_2["rank"] = ad_2.groupby("Area")["Year"].rank("max", ascending=False)
ad_2 = ad_2[ad_2["rank"] == 1]
ad_2.drop(["rank"], axis=1, inplace=True)
```

```python
social_alg = pd.merge(new, ad_2, on=['Area'], how="outer")
social_alg = social_alg.dropna(subset=['Area Code', 'Year_agri', 'aggies', 'Year_poor', 'Poor Percent','ag_rate', 'PPeeps', 'Year','Population', 'GDP  (2015 US$, mil)']).reset_index()
social_alg
```

```python
insecurity = pd.read_csv('../shared_data_read_only/Food Insecurity.csv')
insecurity['Year'] = insecurity['Year'].replace('^\d{4}\D', regex=True)

insecurity['Year'] = insecurity['Year'].str.split('-', expand=True)[1].astype(int)
insecurity["rank"] = insecurity.groupby("Area")["Year"].rank("max", ascending=False)
insecurity = insecurity[insecurity["rank"] == 1]
insecurity.drop(["rank"], axis=1, inplace=True) # want prevalance of udnernoursishment
```

```python
insecurity.drop(['Average fat supply (g/cap/day) (3-year average)', 'Average protein supply (g/cap/day) (3-year average)', 'Area Code', 'Unnamed: 0'], axis=1, inplace=True)
insecurity['Prevalence of undernourishment (percent) (3-year average)'] = insecurity['Prevalence of undernourishment (percent) (3-year average)'].str.replace(r'<', '').astype(float)
insecurity_alc = pd.merge(social_alg, insecurity, on=['Area'], how="outer")
insecurity_alc = insecurity_alc[insecurity_alc['Area Code'].notna()]
insecurity_alc
```

```python
social = insecurity_alc[['Area']]
numerator = insecurity_alc['Poor Percent'] * insecurity_alc['ag_rate'] * insecurity_alc['Prevalence of undernourishment (percent) (3-year average)']
denominator = insecurity_alc['GDP  (2015 US$, mil)'] / insecurity_alc['Population']
social['Social Good Metric'] = numerator / denominator
```

```python
social.dropna()

bop = social.sort_values(by='Social Good Metric', ascending=False).head(5)
plt.figure(figsize=(10,5))
sns.barplot(x="Area",y = 'Social Good Metric', data=bop).set_title(label='Vibes');
```
