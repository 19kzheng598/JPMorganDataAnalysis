```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('shared_data_read_only/Agricultural Employment.csv')
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df
```

```python
df["rank"] = df.groupby("Area")["Year"].rank("max", ascending=False)
```

```python
df
```

```python
df["rank"] 
```

```python
df= df[df["rank"] == 1]
```

```python
df
```

```python
df.drop(["rank"], axis=1, inplace=True)
```

```python
df
```

```python
pf = pd.read_csv('shared_data_read_only/Poverty Rate.csv')
```

```python
pf
```

```python
pf.isnull().sum()
```

```python
pf.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
pf = pf.rename(columns={'Country' : 'Area'})
```

```python
ad = pd.read_csv('shared_data_read_only/Agriculture Dataset.csv')
ad = ad[['Area', 'Year', 'Population']]
ad["rank"] = ad.groupby("Area")["Year"].rank("max", ascending=False)
ad = ad[ad["rank"] == 1]
ad.drop(["rank"], axis=1, inplace=True)
ad
```

```python
new = pd.merge(df, pf, on=['Area'], how="outer")
new = pd.merge(new, ad, on=['Area'], how="outer")
```

```python
new = new.dropna(subset=['Area Code', 'Year_x', 'Number of People Working In Agriculture', 'Year_y', 'Percent of Population Below Poverty Line','Year','Population'])
```

```python
new = new.rename(columns={'Year_x' : 'Year_agri', 'Year_y' : 'Year_poor', 'Year' : 'Year_pop', 'Number of People Working In Agriculture' : 'aggies', 'Percent of Population Below Poverty Line': 'Poor Percent'})
new['ag_rate'] = new['aggies'] / new['Population']
```

```python
new['Poor Percent'] = new['Poor Percent'].str.rstrip("%").astype(float)/100
```

```python
new['PPeeps'] = new['Poor Percent'] * new['Population'] 
new
```

```python
new.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(new.corr(), cbar=True, annot=True, cmap = 'Blues');
```

Correlation between "ag_rate" & "Percent of Population Below Poverty Line" = .33 <br>
Correlation between "aggies --> agriculture workers" & "PPeeps --> Poor People" = 0.94 <br>

```python
plt.figure(figsize=(10,5))
sns.regplot(x="Poor Percent", y="ag_rate", data=new).set_title(label="Correlation between Percent of Poor People vs. Agriculture Employment Rate");
```

Populate above graph with country names for the points

```python
mew = new.copy()
pew = mew.sort_values(by='PPeeps', ascending=False).head(5)
mew = mew.sort_values(by='PPeeps', ascending=True).head(5)
plt.figure(figsize=(10,5))
sns.barplot(x="Area",y = 'PPeeps', data=pew).set_title(label='Top 5 countries with Highest # of Poor People');
```
